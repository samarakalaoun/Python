```python
#Introduction: This game is a simulation of a trip to London. The purpose of the game is for the player to know basic informations about the destination. If the player get the right answers in differet situations, he/she wins and can continue on an Eurotrip.
import time
import sys

def start(l1_chance):
    """Check if the player is the right fit for the game"""
    welcome = input("Do you want to join me on this trip?(Y/N)")
    welcome = welcome.upper()

    if 'Y' == welcome or 'YES' == welcome or 'SURE' == welcome or 'ABSOLUTELY' == welcome:
        print(f"\nGreat! So we\'ll go to London. If I see that you are ready to be a good tourist, I can extend your trip to be an Eurotrip.\n")
        time.sleep(4)
        print("***************************************************************************************************************************")
        time.sleep(1)
        print(f"\nWe are at San Francisco\'s International Airport waiting for our flight to London.")
        time.sleep(3)
        print("It\'s a low-cost airline, so we have to buy our meals in the app.")
        time.sleep(2)
        level_1(l1_chance)
        
    elif 'N' == welcome or 'NO' in welcome or 'NOPE' in welcome or 'I DON\'T'  in welcome or 'I DO NOT' in welcome :
        print ("Too bad! Traveling is the best thing of the world, I\'ll have to find another person.")
        time.sleep(3)
        fail()
    else:
        print("Please, answer \"Yes\" or \"No\".")
        time.sleep(1)
        start(l1_chance)

def level_1(chance):
    """Ask the player if he knows how long is the flight from San Francisco to London."""
    phase_1 = input("How long is the flight?(in hours)")
    
    try:
        """Have an option for the player to try again if he/she does not type the answer in the right format"""
        phase_1 = int(phase_1)
        """Transforming the answer, originally a string, to an integer"""
        if 10 == phase_1:
            print(f"\nThe flight attendant confirmed it, you\'re right! The flight is {phase_1}.")
            time.sleep(3)
            print(f"Thank you for helping me! I\'m so excited to arrive in London.\n")
            time.sleep(3)
            level_2_intro()
        elif 8 <= phase_1 and phase_1 < 10 and phase_1 != 9:
            print(f"\nThe flight attendant just told me that it is a 10 hours flight.\nI guess it will be fine not to have considered {10 - phase_1} hours of flight for my hunger.\n")
            time.sleep(3)
            level_2_intro()
        elif 12 >= phase_1 and 10 < phase_1 and phase_1 != 11:
            print(f"\nThe flight attendant just told me that it is a 10 hours flight.\nI guess it will be fine to have considered {phase_1 - 10} more hours of flight.\n")
            time.sleep(3)
            level_2_intro()
        elif phase_1 == 0:
            print("Seriously? The flight would have ZERO hours?")
            time.sleep(2)
            print("Just try again and play it seriously.")
            time.sleep(2)
            level_1(chance)
        elif phase_1 < 0:
            print("Seriously? The flight would have negative hours?")
            time.sleep(2)
            print("Just try again and play it seriously.")
            time.sleep(2)
            level_1(chance)
        elif phase_1 == 11:
            print(f"\nThe flight attendant just told me that it is a 10 hours flight.\nI guess it will be fine to have considered {phase_1 - 10} more hour of flight.\n")
            time.sleep(3)
            level_2_intro()
                                
        elif phase_1 == 9:
            print(f"\nThe flight attendant just told me that it is a 10 hours flight.\nI guess it will be fine not to have considered {10 - phase_1} hour of flight for my hunger.\n")
            time.sleep(3)
            level_2_intro()
        
        elif phase_1 > 12 and phase_1 <= 20:
            print(f"\nOh no, the flight attendant just told me that the flight is not {phase_1} hours, it\'s actually 10 hours.")
            time.sleep(3)
            print(f"Because of you, I bought food for {phase_1 - 10} hours more than needed. Such a waste of money.")
            time.sleep(4)
            print(f"Sorry, but seems like you won\'t be a good company.\n")
            time.sleep(3)
            fail()
        elif 8 > phase_1 and phase_1 >= 5:
            print(f"\nOh no, the flight attendant just told me that the flight is not {phase_1} hours, it\'s actually 10 hours.")
            time.sleep(3)
            print(f"Because of you, I\'ll either be hungry for {10-phase_1} hours or I\'ll have to spend extra money buying food during the flight.")
            time.sleep(4)
            print(f"Sorry, but seems like you won\'t be a good company.\n")
            time.sleep(3)
            fail()           
            
        elif phase_1 > 20:
            chance = chance - 1
            if chance == 0:
                print("I gave you the chance, you didn\'t know how to use it. I don\'t think we can travel anymore.\n")
                time.sleep(2)
                fail()
            else:
                print(f"You gotta be kidding... We\'re not going to the moon, {phase_1} hours is unreasonable.")
                time.sleep(2)
                print("Try one more time with a plausible number please.")
                level_1(chance)
            
        elif phase_1 < 5:
            chance = chance - 1
            if chance == 0:
                print("I gave you the chance, you didn\'t know how to use it. I don\'t think we can travel anymore.\n")
                time.sleep(2)
                fail()
            else:
                print(f"You gotta be kidding... with {phase_1} hours we can\'t even cross the country. Try one more time, but be more realistic.")
                time.sleep(2)
                print("Try one more time with a plausible number please.")
                level_1(chance)
                            
                        
    except:
        #If the player does not insert the answer in the right format
        print("Don\'t make it complicated, just type the number, simple like that.")
        input("<Press enter to continue>")
        level_1(chance)

def level_2_intro():
    """Story behind level 2"""
    print("***************************************************************************************************************************\n")
    time.sleep(1)
    print("So we are at a restaurant in London.")
    time.sleep(2)
    print("I\'ve ordered a dish, but you were on your phone, so you didn\'t pay attention.") 
    time.sleep(3)
    print("The waiter replied: \"You didn\'t come to London if you didn\'t eat that!\".")
    time.sleep(3)
    level_2(chance_2)
    
def level_2(chance_2):
    """If user knows England's most famous dish"""
    
    phase_2 = input("What did I order?")
    phase_2 = phase_2.upper()
    if 'FISH' in phase_2 and 'CHIP' in phase_2:
        print("\nGreat! It look\'s like you know well about the place we\'re in, I guess you may be a good tourist.\n")
        time.sleep(3)
        level_3_intro()
    else: 
        chance_2 -= 1
        if chance_2 == 0:
            print("I gave you the chance, you didn\'t know how to use it. I don\'t think you\'re able to keep traveling around Europe.\n")
            time.sleep(2)
            fail()
        else:
            print("I know, it\'s not that simple to guess it right on the first time. I\'ll give you one more chance.")
            time.sleep(2)
            level_2(chance_2)
            
def level_3_intro ():
    """Story of level 3, the final round"""
    print("***************************************************************************************************************************\n")
    time.sleep(1)
    print("London is so expensive, we spent more than expected.")
    time.sleep(2)
    print("Now you have to control your budget.")
    time.sleep(2)
    print("I\'ll have to stay in London. But you can continue your trip.")
    time.sleep(2)
    level_3()
           
            
def level_3 ():
    """If the player knows about the possibility of going from London to Paris by bus"""
    phase_3 = input("Is it possible to go from London to Paris by bus?Y/N")
    phase_3 = phase_3.upper()
    if 'Y' == phase_3 or 'YES' == phase_3 or 'SURE' == phase_3 or 'ABSOLUTELY' == phase_3:
        print("\nAmazing! You proved that you can travel by yourself.")
        print("I\'ll just stay in London and you can continue the Eurotrip.")
        print("CONGRATULATIONS!")
        input("<Press enter to exit the game>")
        exit(0)
                
    elif 'N' == phase_3 or 'NO' in phase_3 or 'NOPE' in phase_3 or 'I DON\'T'  in phase_3 or 'I DO NOT' in phase_3 :
        print ("\nOps, you got it wrong. It\'s possible to go by bus on the Channel Tunnel.")
        time.sleep(2)
        print("You\'re still not ready to go to Europe by yourself.")
        time.sleep(2)
        fail()        
    else:
        # gives the player the chance to answer correctly
        print("Please, answer \"Yes\" or \"No\".")
        time.sleep(1)
        level_3()       
    
def fail():
    print("\nGAME OVER")
    input("<Press enter to exit the game>")
    exit(0)
    
eurotrip = "Welcome to the Eurotrip Game!\n"
for letter in eurotrip:
    """Loop to show each letter at a time"""
    print(letter, end="")
    time.sleep(0.1)
time.sleep(1)
print("\nI\'m going to London and I have one extra ticket.")
time.sleep(2)
level1_chance = 2
chance_2 = 2



start(level1_chance)
                
```
